```python
with open('data.txt') as f:
    data = f.read().split('\n')
```

```python
import collections
```

```python
collections.Counter([col for row in data for col in row])
```

```python
from string import ascii_lowercase
```

```python
def get_cart_positions(track):
    positions = {}
    letters = list(ascii_lowercase)
    carts = set(['^', 'v', '<', '>'])
    rows = len(track)
    cols = len(track[0])
    for i in range(rows):
        for j in range(cols):
            cart = track[i][j]
            if cart not in carts:
                continue
            letter = letters.pop(0)
            positions[letter] = {'next_turn': 'left', 'direction': cart, 'position': (i, j)}
    return positions
```

```python
initial_cart_positions = get_cart_positions(data)
```

```python
def build_cartless_track(track, initial_cart_positions):
    new_track = [[y for y in col] for col in track]
    for cart in initial_cart_positions:
        x, y = initial_cart_positions[cart]['position']
        if cart == '^':
            new_track[x][y] = '|' 
        if cart == 'v':
            new_track[x][y] = '|'
        if cart == '<':
            new_track[x][y] = '-'
        if cart == '>':
            new_track[x][y] = '-'
    return [''.join(row) for row in new_track]
```

```python
cartless_track = build_cartless_track(data, initial_cart_positions)
```
