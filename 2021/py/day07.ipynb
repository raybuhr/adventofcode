```python
def find_fuel_cost(data, position):
    dist = np.abs(data - position)
    fuel = [
        np.arange(1, d+1).sum()
        for d in dist
    ]
    return sum(fuel)
```

```python
find_fuel_cost(data, np.mean(data).round(0))
```

```python
fuel_costs = [
    find_fuel_cost(data, i)
    for i in range(max(data))
]
```

```python
sns.scatterplot(x=np.arange(max(data)), y=fuel_costs)
```
